```python
import pandas as pd
import numpy as np
import os
from dotenv import load_dotenv
import quandl
from fredapi import Fred 
import alpaca_trade_api as tradeapi
from dotenv import load_dotenv
from datetime import datetime
```

```python
load_dotenv(" ")
```

```python
alpaca_key = os.getenv("ALPACA_API_KEY")
alpaca_secret = os.getenv("ALPACA_SECRET_KEY")
quandl_key = os.getenv("QUANDL_API_KEY")
fred_key = os.getenv("FRED_API_KEY")
```

```python
timeframe = "1D"

start_date = pd.Timestamp("2015-05-01", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2021-07-10", tz="America/New_York").isoformat()
```

```python
api = tradeapi.REST(
    alpaca_key,
    alpaca_secret,
    api_version = "v2"
)
```

```python
#Equityies Data

tickers = ["SPY","DIA","QQQ"]

equity_data = api.get_barset(
    tickers,
    timeframe,
    start=start_date,
    end=end_date,
    limit=1000
).df

equity_data.columns = ['_'.join(col) for col in equity_data.columns]

equity_data = equity_data[["SPY_close",
                        "DIA_close",
                        "QQQ_close",
                      ]]
equity_data.columns = ["S&P 500", "Dow Jones", "NASDAQ"]
equity_data = equity_data.sort_index(ascending=False)
equity_data.head()
```

```python
data = "NASDAQ"


date = equity_data[data].first_valid_index()
value = equity_data[data].loc[date]
date = '{:%B %d, %Y}'.format(date)
    
print(f"As of {date}, the price of {data} ETF was ${value:.2f}.")
```

```python
#ECONOMIC DATA
economic_data = quandl.get(["FRED/A191RL1Q225SBEA",
                            "FRED/UNRATE", 
                            "FRED/T10YIE", 
                            "FRED/HOUST"],
                           start_date="2010-12-31", 
                           end_date="2021-07-31",
                          )
economic_data.columns = ["GDP","Unemployment Rate","10Yr Inflation", "Housing"]
```

```python
yield_curve = quandl.get("USTREASURY/YIELD", authtoken="m6nQZWVxbd6F2y_ESnBH")
```

```python
new_pd = pd.concat([economic_data,yield_curve['10 YR']], axis="columns")
new_pd.index = pd.to_datetime(new_pd.index)
new_pd = new_pd.resample('Q').mean()
new_pd = new_pd.sort_index(ascending=False)
new_pd.head(10)
```

```python
data = "Unemployment Rate"


date = new_pd[data].first_valid_index()
value = new_pd[data].loc[date]
date = '{:%B %d, %Y}'.format(date)
    
print(f"The latest {data} was reported on {date} and came out to be {value:.2f}.")
```
